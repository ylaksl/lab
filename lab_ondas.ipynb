```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
r_56 = pd.read_excel('osci_excel.xlsx')
r_150 = pd.read_excel('osci_excel_150.xlsx')
```

```python
r_150
```

```python
r_56 = r_56[['frquencia','volts (mV)']]
r_56= r_56[r_56['volts (mV)'] != 0]


r_150 = r_150[['frquencia','volts (mV)']]
r_150= r_150[r_150['volts (mV)'] != 0]
```

```python
R_1 = 55.2 
R_2 = 148.2

c = 1e-6
l = 4.18e-3

w0 = 1/np.sqrt(c*l)
f0 = w0/(2*np.pi)
```

```python
f0 
```

```python
r_56['corrente'] =  round(r_56['volts (mV)']/R_1, 2)
r_150['corrente'] =  round(r_150['volts (mV)']/R_2, 2 )
```

```python
plt.scatter(r_56['frquencia'], r_56['corrente'])
plt.scatter(r_150['frquencia'], r_150['corrente'])
plt.plot(r_56['frquencia'], r_56['corrente'], label = r'$R_{1} = 55.2$ $\Omega$')
plt.plot(r_150['frquencia'], r_150['corrente'], label = r'$R_{2}$ = 148.2 $\Omega$')
plt.axvline(x = f0, c = 'black', label = f'$f_{0}$ = {f0/1000:.2f} kHz')
plt.xscale('log')
plt.xlabel(r"Frequncia (Hz)", family = 'serif')
plt.ylabel(r"Corrente (mA)", family = 'serif')
plt.grid(color = 'gray', linestyle=':')
plt.legend(frameon = False)
plt.tick_params(direction='in', which = 'both', bottom = 1, top =1, left = 1, right = 1)
plt.savefig('freq_ress', dpi = 500)
```
